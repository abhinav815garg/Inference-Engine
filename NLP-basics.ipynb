# **Named Entity Recognition(NER) with NLTK**

```python
from nltk import word_tokenize, pos_tag, ne_chunk
nltk.download('averaged_perceptron_tagger')
nltk.download('maxent_ne_chunker')
nltk.download('words')
sentence = "John is trying to buy 2% shares of Nvidia"
tokens = word_tokenize(sentence)
tokens_pos = pos_tag(tokens)
print(tokens_pos)
```

```python
namedEntity = ne_chunk(tokens_pos)
print(namedEntity)
```

# **Text Summarization using spaCy**

```python
text = "Earth, the third planet from the Sun in our solar system, is a marvel of cosmic happenstance and intricate balance. Spanning a vast 12,742 kilometers in diameter, its rich diversity of landscapes, from sprawling oceans to towering mountains, supports an extraordinary array of life forms. With a surface predominantly covered by water, Earth is often referred to as the \"Blue Planet,\" highlighting its oceans as crucial elements in regulating climate and nurturing biodiversity.Beyond its physical features, Earth is a dynamic system of interconnected processes. The atmosphere, composed mostly of nitrogen and oxygen, shields life from harmful radiation and regulates climate through complex interactions with oceans and land. Its magnetic field protects us from solar winds and cosmic rays, essential for maintaining habitable conditions.Humanity's impact on Earth, however, has become increasingly profound. Industrialization, urbanization, and resource exploitation have strained ecosystems and accelerated climate change, posing significant challenges for future generations.Despite these challenges, Earth remains a resilient planet, capable of regeneration given the right conditions. Efforts in conservation, sustainable development, and global cooperation are crucial in preserving Earth's biodiversity and ensuring a sustainable future for all life forms that call it home. Understanding and respecting the delicate balance of our planet is paramount in securing a harmonious coexistence with nature for generations to come."
textDoc = nlp(text)
```

```python
from string import punctuation
keywords = []
desiredPOS = ['PROPN', 'ADJ', 'NOUN', 'VERB']
for token in textDoc:
  if token.text in nlp.Defaults.stop_words or token.text in punctuation:
    continue
  if token.pos_ in desiredPOS:
    keywords.append(token.text)
print(keywords)
```

```python
from collections import Counter
freq = Counter(keywords)
print(freq)
```

```python
highestFreq = freq.most_common(1)[0][1]
print(highestFreq)
for word in freq:
  freq[word] = freq[word]/highestFreq
print(freq)
```

```python
score = {}
for sentence in textDoc.sents:
  for token in sentence:
    if token.text in freq:
      if sentence in score:
        score[sentence] += freq[token.text]
      else:
        score[sentence] = freq[token.text]
print(score)
```

```python
sortedScore = sorted(score.items(), key=lambda x: x[1], reverse=True)
print(sortedScore)
```

```python
textSummary = []
totalSentences = 4
total = 0
for i in range(len(sortedScore)):
  textSummary.append(sortedScore[i][0])
  total += 1
  if total >= totalSentences:
    break
print(textSummary)
```
